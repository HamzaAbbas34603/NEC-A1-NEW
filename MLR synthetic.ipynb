```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Function to calculate MAPE
def calculate_mape(y_true, y_pred):
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100

# Function to plot scatter diagram
def plot_scatter(y_actual, y_predicted, title):
    plt.scatter(y_actual, y_predicted, alpha=0.5)
    plt.title(title)
    plt.xlabel('Actual Values')
    plt.ylabel('Predicted Values')
    plt.plot([y_actual.min(), y_actual.max()], [y_actual.min(), y_actual.max()], 'k--', lw=2)  # Diagonal line for reference
    plt.show()
```

```python
# Load the synthetic dataset
df_synthetic = pd.read_csv('normalized_A1_synthetic.csv')  # Adjust the path accordingly

# Split the dataset into input features (X) and the target variable (y)
X_synthetic = df_synthetic.iloc[:, :-1]  # Assuming the last column is the target
y_synthetic = df_synthetic.iloc[:, -1]

# Split data into training and testing sets
X_train_synthetic, X_test_synthetic, y_train_synthetic, y_test_synthetic = train_test_split(X_synthetic, y_synthetic, test_size=0.2, random_state=42)

# Create and train the MLR model
mlr_model_synthetic = LinearRegression()
mlr_model_synthetic.fit(X_train_synthetic, y_train_synthetic)

# Make predictions and evaluate the model
predictions_synthetic = mlr_model_synthetic.predict(X_test_synthetic)

# Calculate and output MAPE
mape_synthetic = calculate_mape(y_test_synthetic, predictions_synthetic)
print(f"Synthetic Dataset MLR Model MAPE: {mape_synthetic}%")

# Generate and display scatter plot
plot_scatter(y_test_synthetic, predictions_synthetic, "Scatter Plot for MLR-F Model on Synthetic Dataset")
```
